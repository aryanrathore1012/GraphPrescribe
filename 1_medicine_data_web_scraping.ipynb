```python
try:
    overall_medicine_info=[]
    for alp_link in all_alp_links:
        browser.get(alp_link)
        medicine_layer=browser.find_element(By.CSS_SELECTOR,'#container > div > div > div.style__inner-container___3BZU9.style__product-grid___3noQW.style__padding-top-bottom-12px___1-DPF')
        medicine_classes=medicine_layer.find_elements(By.TAG_NAME,"a")    
        list_medicine_link=[]
        for link in medicine_classes:
            list_medicine_link.append(link.get_attribute("href"))
        all_medicine_info=[]
        for i in range(len(list_medicine_link)):

            medicine_info={}

            medicine_link=list_medicine_link[i]
            browser.get(medicine_link)
            try:
                # getting medicine name
                medicine_name=browser.find_element(By.CLASS_NAME,"DrugHeader__title-content___2ZaPo").text
                medicine_info['name']=medicine_name
            except:
                continue
            try:
                # getting medicine manufacturer
                medicine_manufacture=browser.find_element(By.CLASS_NAME,"DrugHeader__meta-value___vqYM0").text
                medicine_info['manufacturer']=medicine_manufacture
            except:
                continue

            try:
                # getting medicine chemicals
                medicine_salt_div=browser.find_elements(By.CLASS_NAME,'DrugHeader__meta___B3BcU')[1]
                medicine_salt=medicine_salt_div.find_element(By.TAG_NAME,"a").text
                medicine_info['chemicals']=medicine_salt
            except:
                continue

            try:
                # getting medicine uses
                uses=[]
                medicine_uses_list=browser.find_element(By.CSS_SELECTOR,"#uses_and_benefits > div > div > div:nth-child(1) > div > ul")
                medicine_uses=medicine_uses_list.find_elements(By.TAG_NAME, "li")
                for use in medicine_uses:
                    uses.append(use.text)
                medicine_info['uses']=uses
            except:
                continue

            try:
                # getting medicine side effects
                side_effects=[]
                medicine_side_effects_list=browser.find_element(By.CSS_SELECTOR,"#side_effects > div > div > div > div:nth-child(3) > div > ul")
                medicine_side_effects=medicine_side_effects_list.find_elements(By.TAG_NAME, "li")
                for effect in medicine_side_effects:
                    side_effects.append(effect.text)  
                medicine_info['side_effects']=side_effects
            except:
                continue

            try:
                # getting factbox 
                factbox=browser.find_element(By.CSS_SELECTOR,'#fact_box > div.DrugPane__content___3-yrB > div > div')

                fact_key=factbox.find_elements(By.CLASS_NAME,'DrugFactBox__col-left___znwNB')

                fact_value=factbox.find_elements(By.CLASS_NAME,'DrugFactBox__col-right___36e1P')
                factbox_dict={}
                for i in range(len(fact_key)):
                    factbox_dict[fact_key[i].text]=fact_value[i].text
                medicine_info['factbox']=factbox_dict
            except:
                continue

        #     print(medicine_name)
        #     print(medicine_manufacture)
        #     print(medicine_salt)
        #     print(uses)
        #     print(side_effects)
        #     print(factbox_dict)

        #     print("\n\n\n")

            all_medicine_info.append(medicine_info)

        overall_medicine_info=overall_medicine_info+all_medicine_info
        
    df_final=pd.DataFrame(overall_medicine_info)
    df_final.to_csv('web_extracted_data_1mg_60.csv')

except:
    df_final=pd.DataFrame(overall_medicine_info)
    
    # change csv name accordingly
    df_final.to_csv('web_extracted_data_1mg_60.csv')
    
# if you need to see the numberof re4cords already saved then see the len of the df_final saved in the except block

# make remianing 'a' from page 30 and 'g' from page 14 and remaining I from page 12, r from page 12,
```

## getting price, safety advice and interaction information

```python
df_med=pd.read_csv("a_to_z_medicine_data_web_scraping.csv", index_col=0)
df_med.head()
```

```python
# saving the names of all the medicines in the dataframe in form of a list/array/series
med_names=df_med['name']

# setting up the driver 
browser=webdriver.Chrome()
browser.get("https://www.1mg.com")
overall_med_price_record=[]
# getting all the webpages after searching the names by selenium
for name in med_names[889:19037]:
    record={}
    try:
        input_field=browser.find_element(By.ID, "srchBarShwInfo")
        input_field.clear()
        input_field.send_keys(name)
        input_field.submit()

        record["name"]=name

        # extracting the discounted price and actual price in different columns of the dataframe
        try:
            discounted_price=browser.find_element(By.CLASS_NAME,"style__price-tag___B2csA").text
            record["discounted_price"]=discounted_price
        except:
            continue

        try:
            actual_price=browser.find_element(By.CLASS_NAME, "style__discount-price___-Cikw").text
            record["actual_price"]=actual_price
        except:
            continue

    except:
        df_temp=pd.DataFrame(overall_med_price_record)
        df_temp.to_csv("half_data.csv")
        continue
    
    overall_med_price_record.append(record)
# adding info to individual the records

    
```
